# File Location

```python
d = os.path.join("../Data/Training/Training_Data(Resized)/")
classes = ['Glass', 'Metal', 'Paper', 'Plastic', 'Trash']
```

```python
data = {'annotations':{}}
dict= {}
for label in classes:
    directory = os.path.join(d, label)
    for file in os.listdir(directory):
        val = [{'label':label},]
        dict[file]=val
```

```python
data['annotations']=dict
data
```

```python
json_string = json.dumps(data, indent=4)
```

# Save to json file

```python
with open('annotation.json', 'w') as outfile:
    outfile.write(json_string)
```
